```python
import pandas as pd
from pandas import json_normalize
import json
```

```python
# loading JSON file
with open("flights.json", "r", encoding="utf-8") as f:
    all_flights = json.load(f)

# list?
assert isinstance(all_flights, list)
```

```python
all_flights_df = json_normalize(all_flights)
```

```python
all_flights_df.isna().sum()
```

```python
df_departure = json_normalize(
    all_flights,
    meta=['flight_date', 'flight_status'],
    record_prefix='departure.'
)[[
    'departure.airport',
    'departure.timezone',
    'departure.iata',
    'departure.icao',
    'departure.terminal',
    'departure.gate',
    'departure.delay',
    'departure.scheduled',
    'departure.estimated',
    'departure.actual',
    'departure.estimated_runway',
    'departure.actual_runway'
]]
```
